```python
from keras.models import model_from_json
with open('model.json','r') as f:
    model = model_from_json(f.read())
```

```python
model.load_weights('model.h5')
```

```python
model.summary()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
ans=[]
```

```python
def pre(images=5,all=0):
    name=['Cat','Dog']
    lis = os.listdir('images')
    if(all==0):
        images = min(images,len(lis))
    else:
        images = len(lis)
    f,axs = plt.subplots(images,1,figsize=(20,20))
    ans = []
    
    for i in range(images):
        image = cv2.resize(cv2.imread('images/'+lis[i]),(128,128))/255.0
        axs[i].imshow(image)
        ans.append(name[int(round(model.predict(np.array([image]))[0][0]))])
    print(ans)
```

```python
pre(all=1)
```
